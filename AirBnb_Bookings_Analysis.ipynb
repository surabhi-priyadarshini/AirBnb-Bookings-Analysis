## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# 1. Handling Missing Values
# Filling missing values in 'reviews_per_month' with 0
data['reviews_per_month'].fillna(0, inplace=True)

# Filling missing values in 'last_review' with 'Not Specified'
data['last_review'].fillna('Not Specified', inplace=True)

# Dropping rows where 'name' or 'host_name' is missing
data.dropna(subset=['name', 'host_name'], inplace=True)

# 2. Data Transformation
# Converting 'last_review' to datetime
data['last_review'] = pd.to_datetime(data['last_review'], errors='coerce')

# 3. Data Normalization/Standardization
# Normalizing 'price' column
data['price_normalized'] = (data['price'] - data['price'].mean()) / data['price'].std()

# 4. Feature Engineering
# Creating a new feature 'total_reviews' as the sum of 'number_of_reviews' and 'reviews_per_month'
data['total_reviews'] = data['number_of_reviews'] + data['reviews_per_month']

# 5. Handling Outliers
# Remove outliers in 'price' column
q1 = data['price'].quantile(0.25)
q3 = data['price'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr
data = data[(data['price'] >= lower_bound) & (data['price'] <= upper_bound)]

# # Display the cleaned dataset
print(data.head(20))
````

#### Chart - 14 - Correlation Heatmap

````python
# Correlation Heatmap visualization

# Calculating the correlation between price and number of reviews
correlation = data['price'].corr(data['total_reviews'])
print(f'The correlation between price and total reviews is: {correlation}')

# Visualizing the Correlation Using a Scatter Plot:
# Setting the size of the plot
plt.figure(figsize=(10, 6))

# Creating a scatter plot to visualize the relationship
sns.scatterplot(x='total_reviews', y='price', data=data)

# Setting plot labels and title
plt.xlabel('Number of Reviews')
plt.ylabel('Price')
plt.title('Scatter Plot of Price vs. Total Number of Reviews')

# Showing the plot
plt.show()

# Visualizing the Correlation Using a Heatmap:
# Calculating the correlation matrix
correlation_matrix = data[['price', 'total_reviews']].corr()

# Setting the size of the plot
plt.figure(figsize=(8, 6))

# Creating a heatmap to visualize the correlation
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)

# Setting plot title
plt.title('Correlation Matrix')

# Showing the plot
plt.show()
````

##### 1. Why did you pick the specific chart?

A heatmap was chosen here because it effectively communicates the correlation matrix in a visually intuitive manner, facilitating deeper insights into the relationships between variables in the dataset.

##### 2. What is/are the insight(s) found from the chart?

Insights from the Correlation Heatmap:

Correlation Strength: The heatmap shows the correlation coefficient between price and total reviews. If the correlation coefficient is closer to 1, it indicates a strong positive correlation, meaning higher prices tend to correlate with more reviews. Conversely, if it's closer to -1, it suggests a strong negative correlation, where higher prices may result in fewer reviews.

Relationship Clarity: By visualizing the correlation matrix, the heatmap provides clarity on whether there exists a significant relationship between price and total reviews. This understanding can guide decisions related to pricing strategies and customer engagement efforts.

Strategic Insights: Insights from the heatmap can inform strategies such as pricing adjustments, promotional campaigns, or service enhancements aimed at optimizing customer satisfaction and maximizing revenue based on the observed correlation between price and review activity.
